# Le funzioni

## Esempio

La funzione `quadrato` accetta come *argomento* un *numero* e restituisce l'argomento moltiplicato per se stesso.

```c++
double quadrato(double x) {
    return x * x;
}
```

La funzione `quadrato` viene applicata (chiamata, invocata) scrivendone il nome e inserendo tra parentesi l'*argomento*.

L'argomento viene *valutato* prima dell'applicazione della funzione e passato come valore.

La funzione vedr√† una copia dell'argomento come se fosse una sua variabile.

```c++
#include <iostream>
using namespace std;

double risultato;
```

```c++
risultato = quadrato(10);
cout << risultato << endl;
```

```c++
risultato = quadrato(10 + 10);
cout << risultato << endl;
```

```c++
risultato = quadrato(10) + quadrato(5);
cout << risultato << endl;
```

```c++
risultato = quadrato(quadrato(10));
cout << risultato << endl;
```

```c++
for (int i = -3; i < 4; i++)
    cout << i << " -> " << quadrato(i) << endl;
```

```c++
double somma_di_quadrati(double x, double y) {
    return quadrato(x) + quadrato(y);
}
```

```c++
cout << somma_di_quadrati(1 + 2, quadrato(2)) << endl;
```

```c++
long long int fattoriale(int n) {
    if (n < 2)
        return 1;
    return n * fattoriale(n - 1);
}
```

```c++
for (int i = 0; i < 11; i++)
    cout << i << " -> " << fattoriale(i) << endl;
```

La funzione `ripeti` accetta due argomenti, una stringa e un numero.
Restituisce la stringa ripetuta tante volte quanto il numero, separando ogni occorrenza del testo con un a capo.

```c++
#include <string>

/** Ripete una stringa
  * @param testo Il testo da ripetere
  * @param ripetizioni Il numero di ripetizioni del testo
  * @return il testo ripetuto \var ripetizioni volte
  */
string ripeti(string testo, int ripetizioni) {
    string ret = "";
    for (int i = 0; i < ripetizioni; i++)
        ret += testo + '\n';
    return ret;
}
```

```c++
for (int i = 0; i < 6; i++) {
    cout << ripeti("Ciao", i) << endl;
    cout << " === " << endl;
}
```
