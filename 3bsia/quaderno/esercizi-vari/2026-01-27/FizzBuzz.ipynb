# FizzBuzz

Scrivi un programma che stampa tutti i numeri da 1 a 100, con due eccezioni. Per i numeri divisibili per 3, stampa "Fizz" al posto del numero, e per i numeri divisibili per 5 (e non per 3), stampa "Buzz".

Quando il programma funziona, modifica il programma in modo che stampi "FizzBuzz" per i numeri divisibili sia per 3 che per 5 (e stampi comunque "Fizz" o "Buzz" per i numeri divisibili solo per uno di questi).

(da Eloquent JavaScript)

```c++
#include <string>
#include <iostream>
using namespace std;

for (int numero = 1; numero < 100; numero++) {
    bool div3o5 = false;
    if (numero % 3 == 0) {
        div3o5 = true;
        cout << "Fizz";
    }
    if (numero % 5 == 0) {
        div3o5 = true;
        cout << "Buzz";
    }
    if (!div3o5)
        cout << numero;
    cout << endl;
}
```
