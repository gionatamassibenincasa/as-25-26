# Chessboard

Scrivi un programma che genera una stringa che rappresenta una schacchiera 8x8.
Usa i caratteri di "a capo" per separare le linee, uno spazio per rappresentare le caselle bianche e un cancelletto per rappresentare quelle nere.

## Esempio

```txt
 # # # #
# # # # 
 # # # #
# # # # 
 # # # #
# # # # 
 # # # #
# # # #
```

Dopo aver scritto un programma che genera la scacchiera 8x8, modificalo per avere una variabile `dim = 8` e cambia il programma in modo tale da farlo funzionare per ogni dimensione della scacchiera.

````c++
#include <string>
#include <iostream>
using namespace std;

const int dim = 8;
````

````c++
string scacchiera = "";

int i = 0;
while (i < dim) {
    int j = 0;
    while (j < dim) {
        if((i + j) % 2 == 0)
            scacchiera += ' ';
        else
            scacchiera += '#';
        j++;
    }
    scacchiera += '\n';
    i++;
}

cout << scacchiera;
````
