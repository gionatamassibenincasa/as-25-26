# Un triangolo di cancelletti

Scrivi il codice C++ che produce in output il seguente "triangolo":

```txt
#
##
###
####
#####
######
#######
```

La prima riga ha un solo cancelletto e un `newline`, la seconda due cancelletti e cos√¨ via fino all'ultima che ha 7 cancelletti.

````c++
#include <string>
#include <iostream>
using namespace std;

int contatore;
````

## Senza strutture iterative

Scrivi il codice senza usare le strutture iterative

````c++
cout << "#" << endl;
cout << "##" << endl;
cout << "###" << endl;
cout << "####" << endl;
cout << "#####" << endl;
cout << "######" << endl;
cout << "#######" << endl;
````

## Costrutto while

Usa solo il costrutto iterativo while

````c++
// Versione "while"
contatore = 1;
while (contatore < 8) {
    string s = "";
    int i = 0;
    while (i < contatore) {
        s += "#";
        i++;
    }
    cout << s << endl;
    contatore++;
}
````

## Costrutto do/while

Usa solo il costrutto iterativo do/while

````c++
contatore = 1;
do {
    string s = "";
    int i = 0;
    do {
        s += "#";
        i++;
    } while (i < contatore);
    cout << s << endl;
    contatore++;
} while (contatore < 8);
````

## Costrutto for

Usa solo il costrutto iterativo for

````c++
for (contatore = 1; contatore < 8; contatore++) {
    string s = "";
    for (int i = 0; i < contatore; i++)
        s += "#";
    cout << s << endl;
}
````
