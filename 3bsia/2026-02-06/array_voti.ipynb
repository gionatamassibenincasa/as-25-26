```c++
/* Esempio: un array di voti */
// Visualizzare con PythonTutor

#include <iostream>
using namespace std;

int main() {
  const double voti[] {7, 7.5, 6, 4, 3, 8, 7.5, 6};
  const int nVoti = sizeof(voti)/sizeof(double);
  
  double somma = 0.;
  for (int i = 0; i < nVoti; i++) {
    // La & Ã¨ usata solo per motivi di visualizzazione
    // su PythonTutor
    const double &voto = voti[i];   
    somma += voto;
  }
    
  cout << "Media: " << somma / nVoti << endl;

  return 0;
}
```
